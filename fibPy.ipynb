```python
import timeit
import matplotlib.pyplot as plt

time_list = []
fibN_list = []

mysetup = '''
t1 = 0 
t2 = 1
tn = 0
def calFib (n,t1,t2,tn):
    for i in range (3,n+1):
        tn = t1 + t2
        t1 = t2
        t2 = tn       
'''
mycode = '''calFib(n,t1,t2,tn)'''


for i in range (1,120):
    codestr = mycode.replace("n", str(i), 1)
    time = timeit.timeit(setup = mysetup, 
                    stmt = codestr, 
                    number = 30000)
    time_list.append(time)
    fibN_list.append(i)
    
plt.plot(fibN_list, time_list)
plt.xlabel('Fibonacci N')
plt.ylabel('Tiempo de ejecucion')
plt.show()
```
