```python
import time
import matplotlib.pyplot as plt

N_list =[]
T_list = []

n = 60
t1 = 0 
t2 = 1
tn = 0

start = time.time()
N_list.append(t1)
end = time.time()
T_list.append(end-start)

start = time.time()
N_list.append(t2)
end = time.time()
T_list.append(end-start)

for i in range (3,n+1):
    start = time.time()
    tn = t1 + t2
    N_list.append(tn)
    t1 = t2
    t2 = tn
    end = time.time()
    T_list.append(end-start)
    
plt.plot(N_list,T_list)
plt.xlabel('Fibonacci N')
plt.ylabel('Tiempo de ejecucion')
```
